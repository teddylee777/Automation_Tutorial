## 마우스 좌표 확인

```python
mouse_position = position()
x = mouse_position.x
y = mouse_position.y

x, y
```

## 마우스 이동

```python
moveTo(x, y, duration=3) 
```

## 마우스 클릭

```python
click(x=x, y=y, clicks=5, interval=2, button='left')
```
